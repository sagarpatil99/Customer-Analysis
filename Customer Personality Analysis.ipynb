# Feature Engineering

```python
# Adding customer age
import datetime as dt
df['Age'] = 2015 - df.Year_Birth
```

```python
# Counting the months of customers registerd to the company
df['Dt_Customer'] = pd.to_datetime(df['Dt_Customer'])
df['Month_Customer'] = 12.0 * (2015 - df.Dt_Customer.dt.year ) + (1 - df.Dt_Customer.dt.month)
```

# Feature Transformation

```python
# selecting the cols to use
subset = df[['Income','Kidhome','Teenhome','Age','Marital_Status','Education']]
print('This is the data we will use for clustering:')
subset.head()
```

```python
pip install yellowbrick
```

```python
# for manupulating time and date columns
from datetime import date, datetime 

# For handling categorical column and scaling numeric columns
from sklearn.preprocessing import OneHotEncoder, OrdinalEncoder, StandardScaler 

# Libraries for clustering and evaluation
from sklearn.cluster import KMeans
from yellowbrick.cluster import KElbowVisualizer
from sklearn.metrics import silhouette_score

import warnings
# To prevent kernel from showing any warning
warnings.filterwarnings("ignore") # To prevent kernel from showing any warning

# set the color palette
palette = sns.color_palette(["#292859",'#373482','#504DB6','#5B59DD'])
sns.palplot(palette) # print color palette
```

```python
# Importing essential libraries for bulding pipeline
from sklearn.pipeline import make_pipeline
from sklearn.compose import ColumnTransformer
```

```python
# examin the min, max and mean of your columns
subset.describe()
```

```python
# I am not scaling the kidhome, teenhome cols, cause there min, max lies between 0 & 2
num_cols = ['Income','Age']
numeric_pipeline = make_pipeline(StandardScaler())
```

```python
ord_cols = ['Education']
ordinal_pipeline = make_pipeline(OrdinalEncoder(categories=[['Graduation', 'Master', '2n Cycle', 'PhD', 'Basic']]))
```

```python
nom_cols = ['Marital_Status']
nominal_pipeline = make_pipeline(OneHotEncoder())
```

```python
# stack your pipelines in column transformer
transformer = ColumnTransformer(transformers=[('num',numeric_pipeline,num_cols),
                                           ('ordinal', ordinal_pipeline,ord_cols),
                                              ('nominal' ,nominal_pipeline,nom_cols)
                               ])
```

```python
transformer
```

```python
# Now lets fit and transform the data
transformed = transformer.fit_transform(subset)
print('Data has been Transformed')
```

# Machine Learning Model

# K-Means Clustering & Cluster's Analysis

### Optimum Clusters using Elbow method

```python
plt.figure(figsize=(12, 8))
elbow_graph = KElbowVisualizer(KMeans(random_state=43), k=10)
elbow_graph.fit(transformed)
elbow_graph.show()
```

##### Disortation score quantifies the average squared distance between each data point in a cluster and its centroid. The distortion score is commonly used as a criterion to evaluate the quality of clustering and to determine the optimal number of clusters.
##### Disortation score = 3415.724 and K = 5 , there are 4 clusters

```python
# using k-means to form clusters
kmeans = KMeans(n_clusters=4, random_state=42)

#fit the data and adding back clusters to the data in clusters column
subset['Clusters'] = kmeans.fit_predict(transformed) 
```

# Model Building

### Building multi classification model to predict the cluster for particular customer

```python
# import required libraries
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.ensemble import GradientBoostingClassifier
```

```python
# separate features and target column
x = subset.drop('Clusters', axis=1)
y = subset['Clusters']
```

```python
# create train and test data
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.3, random_state=42)
```

```python
# adding GradientBoostingClassifier to transformer pipeline
final_pipeline = make_pipeline(transformer, GradientBoostingClassifier())
```

```python
# fit the data to new pipeline & model
final_pipeline.fit(x_train, y_train)
```

```python
# check the accuracy of our model
final_pipeline.score(x_test, y_test)
```
